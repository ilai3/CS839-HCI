```python
import numpy as np    # linear algebra 
import pandas as pd    # Data processing, Input & Output load    
import matplotlib.pyplot as plt    # Visualization & plotting
from sklearn.preprocessing import LabelEncoder
import warnings   
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import pickle
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('healthcare-dataset-stroke-data.csv')
data.head()
```

```python
data['bmi'].fillna(data['bmi'].mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
data.drop('id',axis=1,inplace=True)
```

```python
enc=LabelEncoder()
gender=enc.fit_transform(data['gender'])
smoking_status=enc.fit_transform(data['smoking_status'])
work_type=enc.fit_transform(data['work_type'])
Residence_type=enc.fit_transform(data['Residence_type'])
ever_married=enc.fit_transform(data['ever_married'])
data['work_type']=work_type
data['ever_married']=ever_married
data['Residence_type']=Residence_type
data['smoking_status']=smoking_status
data['gender']=gender
data
```

```python
le = LabelEncoder()
encoded_work_type = le.fit_transform(data['smoking_status'])

print("{:<15} {}".format('Work Type', 'Encoded Value'))
print("-----------------------")
for i in range(len(data['smoking_status'])):
    print("{:<15} {}".format(data['smoking_status'][i], encoded_work_type[i]))
```

```python
X=data.drop('stroke',axis=1)
Y=data['stroke']
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
std=StandardScaler()
X_train_std=std.fit_transform(X_train)
X_test_std=std.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train_std,Y_train)
Y_pred=dt.predict(X_test_std)
with open('DTModel.pkl', 'wb') as f:
    pickle.dump(dt, f)
ac_dt=accuracy_score(Y_test,Y_pred)
ac_dt
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train_std,Y_train)
Y_pred_lr=lr.predict(X_test_std)
with open('LRModel.pkl', 'wb') as f:
    pickle.dump(lr, f)
ac_lr=accuracy_score(Y_test,Y_pred_lr)
ac_lr
```

```python
from sklearn.svm import SVC
sv=SVC()
sv.fit(X_train_std,Y_train)
Y_pred=sv.predict(X_test_std)
with open('SVCModel.pkl', 'wb') as f:
    pickle.dump(sv, f)
ac_sv=accuracy_score(Y_test,Y_pred)
ac_sv
```
